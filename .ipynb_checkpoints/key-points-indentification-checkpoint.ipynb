## Image Features Extraction and Creation of Classes

Rather than working on the large dataset directly, I will first work on few individual images for this task, test it and later on merge the process with dataset from a different source.

```python
all_files = []

hands = []
nothands = []

types = ('images/hands/*.jpg', 'images/not-hands/*.jpg')

for files in types:
    all_files.extend(glob.glob(files))

for image_path in all_files:
    if 'not-hands' in image_path:
        nothands.append(image_path)
    else:
        hands.append(image_path)
```

Resizing the images makes it better for to work on with the upcoming process. On testing with few images, I found out thet images of size 64 X 64 are reasonably distinguishable if it contains a hand or not.

```python
def resize_images(img):
    image_resized = resize(img, (64, 64), anti_aliasing=True)
    return image_resized
```

```python
index = np.random.randint(0,len(hands))
tst = hands[index]
img = mpimg.imread(tst)
image_resized = resize_images(img)
plt.imshow(image_resized)
```

### Spatial Bin Features:


```python
def get_spatialbin_features(img):
    features = resize_images(img).ravel() 
    return features

get_spatialbin_features(img).shape
```

### HOG (Histogram of Oriented Gradients) Features

```python
def get_hog_features(img):
    img = resize_images(img)
    hist = feature.hog(img, orientations=5, pixels_per_cell=(2,2),cells_per_block=(2, 2), block_norm='L2-Hys')
    return hist

get_hog_features(img).shape
```

#### Combination and Vectorization

```python
def get_image_features(img):
    combined_features = []
    
    spatial_bin_features = get_spatialbin_features(img)
    combined_features.append(spatial_bin_features)
        
    hog_features = get_hog_features(img)
    combined_features.append(hog_features)
        
    return (np.concatenate(combined_features))


def extract_and_tag_features(paths):
    all_features = []
    for path in paths:
        image = mpimg.imread(path)
        all_features.append(get_image_features(image))
    return all_features
```

```python
hand_features = extract_and_tag_features(hands)
nothand_features = extract_and_tag_features(nothands)

print(hand_features[1].shape)
print(nothand_features[1].shape)
```

## Modelling and Classification

Following the same process I have created pickled objects of images of the features of hands (hand_features.pkl)  and not-hands (nothand_features.pkl) for a large number of images which I will now use in order to create the classifiers.

```python
with open("hand_features.pkl", "rb") as file_obj:
    xf = pickle.load(file_obj)
with open("nothand_features.pkl", "rb") as file_obj:
    yf = pickle.load(file_obj)
```

```python
x = np.vstack((xf, yf)).astype(np.float64)
y = np.hstack((np.ones(len(xf)), np.zeros(len(yf))))
x.shape, y.shape
```

```python
#Splitting Train, Test and Validation Data

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=1)
x_train, x_valid, y_train, y_valid = train_test_split(x_train, y_train, test_size=0.20, random_state=1)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

#### Linear SVM

```python
x_scaler = StandardScaler().fit(x_train)

x_train = x_scaler.transform(x_train)
x_valid = x_scaler.transform(x_valid)
x_test = x_scaler.transform(x_test)

linear_svm_clf = svm.LinearSVC()
linear_svm_clf.fit(x_train, y_train)
print(linear_svm_clf.score(x_valid, y_valid))
```

#### Deep Neural Networks

```python
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, InputLayer
from keras import backend as K
```

```python
batch_size = 16
num_classes = 2
epochs = 10

y_train_keras = keras.utils.to_categorical(y_train, num_classes)
y_valid_keras = keras.utils.to_categorical(y_valid, num_classes)
y_test_keras = keras.utils.to_categorical(y_test, num_classes)
dnn_model = Sequential()
dnn_model.add(InputLayer(input_shape=(x_train.shape[1],)))
dnn_model.add(Dense(100, activation='elu'))
dnn_model.add(Dropout(0.4))
dnn_model.add(Dense(24, activation='elu'))
dnn_model.add(Dense(2, activation='sigmoid'))
dnn_model.summary()
```

```python
dnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
hist = dnn_model.fit(x_train, y_train_keras, batch_size, epochs, validation_data=(x_valid, y_valid_keras))
```

## Key Points Identification to Explore the Points in a Hand

```python
import imutils
import cv2
import numpy as np
import random as rnd
```

### 1. Specification of Masks in an Image Based on the Skin Colour Range

```python
def get_mask(image):
    # BGR2YCR CBskin colour range 
    lower_bound = np.array([0,133,77], dtype=np.uint8)
    upper_bound = np.array([235,173,127], dtype=np.uint8)

    mask = cv2.inRange(image, lower_bound, upper_bound)
    #removing the noise 
    kernelOpen = np.ones((4, 4))
    kernelClose = np.ones((10, 10))

    maskOpen = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernelOpen)
    maskClose = cv2.morphologyEx(maskOpen, cv2.MORPH_CLOSE, kernelClose)
    final_mask = maskClose
    return final_mask
```

### 2. Getting the Convex Hull

```python
def get_convex_hull(contours):
    hull_list = []
    for i in range(len(contours)):
        hull_list.append(cv2.convexHull(contours[i]))
    return hull_list
```

### 3. Contour Extraction

```python
def get_contours(img):
    contours= cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    return contours
```

#### In order to proceed further, and create functions (drag-and-drop) based on the directional movement of these two points, a simple algorithm would be:

1. Track the movement of these points and match the correspondence with in a range of an object (suppose 'A') on screen
2. If distance between to points is sufficiently small and both the points lie within specified area range of object A then:
    
    - Catch the object
    
    - If the points are moving thereafter, drag object A in the direction of their movement
        
        i) Keep dragging until condition 'ii' is satisfied.
        
        ii) If the distance between to points becomes sufficiently large, release object A at that point.





#### Possible Extentions of the Approach:

Similar algorithms can be devised based on the movement of such points. In the same way, generation of multiple points, rather than only two, would be more beneficial in making such functions more effictive.
